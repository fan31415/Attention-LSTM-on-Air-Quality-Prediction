```python
# Shared this network parameter only within shared weather station datas and air qualities datas

# Two layer lstm model
# The shape of inputs is [batch_size, num_steps, feature_num]
# batch index: current number in range(batch_count)
# state: only pass when batch index is not 0, it is the final state of last batch
# by this we can pass state between batches

        
def lstm_layer(inputs, batch_index, state, num_step = NUM_STEPS, batch_size = BATCH_SIZE, state_size = 300, layer_num = 2):
    print("inputs' shape: " + str(inputs.get_shape()))
    with tf.variable_scope("deep_lstm_model"):
        
        
        stacked_cell = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.LSTMCell(state_size) \
                                                   for _ in range(layer_num)])
        
        if batch_index == 0:
            initial_state = stacked_cell.zero_state(batch_size, dtype=tf.float32)
        else:
            initial_state = state
                
        # inputs' shape is [Time_Step, TimeCount, Feature_Number]
        # outputs' shape is [batchz_size, num_timestep, state_size], final_state's shape is [batch_size, state_size]
        outputs, final_state = tf.nn.dynamic_rnn(stacked_cell, \
                                            inputs, initial_state=initial_state, dtype = tf.float32)
        # We get output of every num steps 
        # We only need the last output of every num step here
        output = outputs[:, -1, :]
        
    return output, final_state


    
```

```python
class LSTM_model(object):
    def __init__(self, batch_size = BATCH_SIZE, state_size = LSTM_HIDDEN_SIZE, layer_num = 2, num_steps = NUM_STEPS,\
                feature_num = AIR_FEATURE_NUM):
        stacked_cell = tf.nn.rnn_cell.MultiRNNCell([tf.nn.rnn_cell.LSTMCell(state_size) \
                                                   for _ in range(layer_num)])
        
        self.inputs = tf.placeholder(tf.float32, [batch_size, num_steps, feature_num])
        
        self.init_state = stacked_cell.zero_state(batch_size, dtype=tf.float32)
        
        state = self.init_state
        
        print(self.inputs.get_shape())
        outputs, state = tf.nn.dynamic_rnn(stacked_cell, \
                                            self.inputs, initial_state= state, dtype = tf.float32)
        
#         print("lstm ", outputs)
        self.final_state = state
        self.output = outputs[:, -1, :]
```

```python
# inputs will be shape [time_count, feature_number]
def generate_air_quality_lstm_data(inputs, batch_size, num_steps):
    
    columns = list(inputs)
    feature_num = len(columns)
    
    row_count = inputs.shape[0]
    
    take_count = row_count - 1
    

    
    # Adjust to make Y one hour later than X inherently
    
    # The last time data will be one of our predictions, so will not include in our input_X
    # To make input X has proper shape, I start the index from n-1
    input_X = inputs.iloc[- take_count - 1: - 1]
    
    # we only need the label data to be our labels
    data_Y = inputs[Labels]
    # predict data will be start after lstm feedforwad through the first num steps data
    # Y start will be one hour later than X to be our label
    Y = data_Y.iloc[- take_count : ]
    
    
        
    
    # Normalized Data
    scaler = StandardScaler()
    input_X = scaler.fit_transform(input_X)
    
    
    #     Arrange X into sequence list    
    sequence_X = []

    sequence_Y = []
    
    for i in range(take_count):
        if i > take_count - num_steps:
            break
        sequence_X.append(input_X[i:i + num_steps])
            # Y start already earlier than X one hour, so here we should minus 1, it will get the correspond Y for X
        sequence_Y.append(Y.values[i+num_steps - 1])
          
            
            
    # Make Batch
    #batch_size is the actuall training records' batch size
    sequence_count = (take_count - batch_size + 1)
    batch_count = sequence_count // batch_size
    actuall_count = batch_count * batch_size
    
    # clip the margin data
    
    sequence_X = sequence_X[-actuall_count:]
    sequence_Y = sequence_Y[-actuall_count:]
    
    
    X_batches = np.split(np.array(sequence_X), batch_count)
          
    Y_batches = np.split(np.array(sequence_Y), batch_count)
    
    

    return X_batches, Y_batches
    
```

```python
# For each station, we build a whole model and train it
def main_model(local_input, X, y, is_training):
    
    local_locations = tf.placeholder(tf.float32, shape = (None, 2))
    with tf.variable_scope("location_fc"):
        local_locations_output = location_FC(local_location)
        
    
    with tf.variable_scope("local_lstm"):
        local_lstm_output = lstm_model()
    
    # For one local
    
    # FNN-RNN hybrid layer for shared stations feature
    
    # What is the dimension of FNN output
    location_FNN_outputs = [None]*STATION_NUM
    # all station use one FNN to simplify training
    sharedFNN = build_location_FNN(layer_num = 3)
    
    # [STATION_NUM]
    for i in range(STATION_NUM):
        # stationAddressVector: eg. (distance, direction)
        shared_FNN_outputs[i] = sharedFNN(inputs = stationAddressVector)
    sharedLSTM = build_shared_LSTM(hidden_num = STATION_NUM)
    #station timed feature means the dynamic feature in each station
    sharedInput = concatStationFeature(stationsTimedFeatures)
    h = [None]*STATION_NUM
    h = sharedLSTM(inputs = sharedInput)
    
    
    
    shared_fc_output = fc_layer(inputs = [shared_FNN_outpus, h], layer_num = 2)
    
    # FNN-RNN hybrid layer for local station feature
    
    # Is this needed for local?
    local_FNN = build_location_FNN(layer_num = 3)
    local_FNN_output = local_FNN(inputs = localAddress)
    
    local_LSTM = build_basic_LSTM(hidden_num = 5)
    local_LSTM_output = local_LSTM(inputs = localTimedFeatures)
    
    local_fc_output = fc_layer(inputs = [local_FNN_outpu, local_LSTM_output], layer_num = 2)
    
    # Use Attention model to adjust shared features
    
    a = attentionNet(inputs = [local_fc_output, shared_fc_output])
    
    scores = getScoresVector(a)
    
    attention_output = shared_fc_output * scores
    
    # Fusion layer to combine local feature and attention model result of shared feature
    fusion_output = fusion_layer(inputs = [local_fc_output, attention_output], layer_num = 1)
    
    # Prediction layer to predict the next day air condition
    preds = pred_weight * fusion_output + pred_bias
    
    #Loss function
    loss = mse(local_air_condition[1:], preds)
    
    train(adam, loss) 
```

```python
def run_eval(sess, test_X, test_Y):
    with tf.variable_scope("model"):
        prediction, _, _ = main_model(train_X, [0.0], False)
    preds = []
    labels = test_Y
    for i in range(TESTING_EXAMPLE_NUM):
        pred = sess.run(prediction)
        preds.append(pred)
        
    rmse = nq.sqrt((preds - labels) ** 2).mean(axis = 1)
    
    print("Mean Square Error: " + str(rmse))
```
